```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.by import By
import pandas as pd
path="/usr/bin/chromedriver"
```

```python
driver_kun=webdriver.Chrome()
driver_kun.get("https://kun.uz/news/list")
```

```python
block_link=[]
try:
    sana=2
    index=1
    while True:
        link_kun = driver_kun.find_element(By.XPATH,'//div[@class="load-more"]').click()
        print("clicked")
        if sana%2 ==0:
            links_kun=[]
            print("olingan boshi")
            elems_kun = driver_kun.find_elements(By.XPATH,"//a[@href]")
            print(len(elems_kun))
            #print(elems_kun)
            for elem in elems_kun[::-1]:
                urls=elem.get_attribute("href")
                print(urls)
                block_link.append(urls)
                break

            
        sana=sana+1
        if(len(block_link)>10):
            break
        time.sleep(2)
except Exception as ex:
    print(ex," katta trydan chiqti")
```

```python
import requests
from bs4 import BeautifulSoup
import os
import pandas as pd
```

```python
links_kun=[]
for url in block_link:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    quotes = soup.find('div', class_="daily-news max-w").find_all(href=True)
    for quote in quotes:
    #     print(quote)
        link = "https://kun.uz"+quote['href']
        print(link)
        links_kun.append(link)
```
